```python
#copiada função apresentado durante a aula

def identificar_outliers(df, coluna):
    # Calculando o primeiro e terceiro quartis
    Q1 = df[coluna].quantile(0.25) #Primeiro Quartil
    Q3 = df[coluna].quantile(0.75) #Terceiro Quartil
    
    # Calculando o IQR (Intervalo Interquartil)
    IQR = Q3 - Q1
    
    # Calculando os limites superior e inferior para identificar os outliers
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    
    # Identificando outliers na coluna especificada
    outliers = df[(df[coluna] < limite_inferior) | (df[coluna] > limite_superior)]
    
    return outliers
```

```python
# pedir para mostrar os outliers
outliers_encontrados = identificar_outliers(df,'Credit amount')

outliers_encontrados
```

```python
#Atenção, feito pelo chatgpt

#resolvi substituir os valores que estão saíndo como outliers, pelo valor da mediana

def substituir_outliers_por_mediana(df, coluna):
    outliers = identificar_outliers(df, coluna)
    mediana = df[coluna].median()
    df.loc[outliers.index, coluna] = mediana
```

```python
substituir_outliers_por_mediana(df, 'Credit amount')
```

```python
#Trouxe os valores e boxplot novamente, mas não fiquei satisfeito ainda, pois os valores estão além do quadrado (2° e 3° quartil)
df.boxplot(column='Credit amount')
plt.title('Coluna')
plt.ylabel('Valores')
plt.show()
```
